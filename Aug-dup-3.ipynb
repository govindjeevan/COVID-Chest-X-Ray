```python
import numpy as np
import cv2
from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.layers import Dense
from keras.preprocessing.image import ImageDataGenerator
from matplotlib import pyplot
```

```python
# create data generator for horizontal shift
datagen = ImageDataGenerator(width_shift_range=[-100,100])
```

```python
#create the iterator
it = datagen.flow_from_directory('val-dup',
target_size = (512, 512),
class_mode = 'binary', batch_size=40)
```

```python
#plot and save in Augmented/0 directory
for i in range(3):
    for j in range(0,40):
        batch = it.next()
        name = 'Augmented-Val/0/'+str(i)+str(j)+'-hs.jpg'
        rgbimg = cv2.cvtColor(batch[0][j].astype('uint8'), cv2.COLOR_BGR2RGB)
        cv2.imwrite(name,rgbimg)
    image = batch[0][0].astype('uint8')
    pyplot.imshow(image)
# show the figure
pyplot.show()
```

```python
# create data generator for vertical shift
datagen = ImageDataGenerator(height_shift_range=0.5)
```

```python
it = datagen.flow_from_directory('val-dup',
target_size = (512, 512),
class_mode = 'binary', batch_size=40)
```

```python
for i in range(1):
    for j in range(0,40):
        batch = it.next()
        name = 'Augmented-Val/0/'+str(i)+str(j)+'-vs.jpg'
        rgbimg = cv2.cvtColor(batch[0][j].astype('uint8'), cv2.COLOR_BGR2RGB)
        cv2.imwrite(name,rgbimg)
    image = batch[0][0].astype('uint8')
    pyplot.imshow(image)
# show the figure
pyplot.show()
```

```python
# create data generator for brightness
datagen = ImageDataGenerator(brightness_range=[0.2,1.0])
```

```python
it = datagen.flow_from_directory('val-dup',
target_size = (512, 512),
class_mode = 'binary', batch_size=40)
```

```python
for i in range(1):
    for j in range(0,40):
        batch = it.next()
        name = 'Augmented-Val/0/'+str(i)+str(j)+'-br.jpg'
        rgbimg = cv2.cvtColor(batch[0][j].astype('uint8'), cv2.COLOR_BGR2RGB)
        cv2.imwrite(name,rgbimg)
    image = batch[0][0].astype('uint8')
    pyplot.imshow(image)
# show the figure
pyplot.show()
```

```python
# create data generator for random zoom
datagen = ImageDataGenerator(zoom_range=[0.5,1.0])
```
